```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
df = pd.read_csv("/content/insurance.csv")
```

```python
df.head()
```

```python
X = df.drop(['charges'], axis=1)
y = df['charges']
```

```python
X_encoded = pd.get_dummies(X, columns=['sex', 'smoker', 'region'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Building the ANN model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Training the model
history = model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Evaluating the model
loss = model.evaluate(X_test_scaled, y_test)
print("Test Loss:", loss)
# Take user input for features
age = float(input("Enter age: "))
bmi = float(input("Enter BMI: "))
children = int(input("Enter number of children: "))
sex_male = int(input("Is the person male? (1 for Yes, 0 for No): "))
smoker_yes = int(input("Is the person a smoker? (1 for Yes, 0 for No): "))
region_northwest = int(input("Is the region Northwest? (1 for Yes, 0 for No): "))
region_southeast = int(input("Is the region Southeast? (1 for Yes, 0 for No): "))
region_southwest = int(input("Is the region Southwest? (1 for Yes, 0 for No): "))

# Create a DataFrame with user input
user_input = pd.DataFrame({
    'age': [age],
    'bmi': [bmi],
    'children': [children],
    'sex_male': [sex_male],
    'smoker_yes': [smoker_yes],
    'region_northwest': [region_northwest],
    'region_southeast': [region_southeast],
    'region_southwest': [region_southwest]
})

# Scale the user input
user_input_scaled = scaler.transform(user_input)

# Predict charges using the trained model
predicted_charge = model.predict(user_input_scaled)

# Print the predicted charge
print("Predicted insurance charge:", predicted_charge[0][0])
```
